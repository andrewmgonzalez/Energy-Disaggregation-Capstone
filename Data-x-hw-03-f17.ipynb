
#### 11) Change the code so that it accepts a list of any stock symbols, ie ['NKE', 'APPL', 'DIS', ... ] and creates a correlation matrix for the time period of the past 100 days


```python
user_input = ['NKE', 'AAPL', 'DIS', 'GOOG']
n=100

df = pd.DataFrame(index = range(0,n), columns = user_input)
for item in user_input:
    df_temp = pd.read_csv('https://www.google.com/finance/historical?output=csv&q=' + item)
    df_temp.index = pd.to_datetime(df_temp['Date'])
    df_temp = df_temp.head(n)
    df_temp = df_temp['Close']
    df[item] = list(df_temp)
    df.index = df_temp.index
print('LAST' ,n , 'DAYS:\n',df.corr())
```
